```python
#Make Scraper below pull all data recuersively, not only that of 'Chosen Stock'
#Rather than recursion, make ui and add start and stop buttons, where script repeats until stop is pressed, also displaying iterations

hist_pop = []
hist_price = []
hist_date = []

def price_fetch (x):
    pr = rh.ask_price(x)
    pr = pr[0]
    return(pr[0])

for i in range(5761):
    ct = str(datetime.datetime.now())
    pf = float(price_fetch(chosen_stock))
    current_price = [ct, pf]
    current_pop = [ct, rh.get_popularity(chosen_stock)]
    
    hist_pop.append(rh.get_popularity(chosen_stock))
    hist_price.append(pf)
    hist_date.append(ct)
    
    #CHANGE 24 TO TICKS REQUIRED FOR 24 HRs
    if i % 100 == 0:
        wr = [hist_date, hist_price, hist_pop]
        zwr = zip(*wr)

        ext = '.csv'
        base = 'stock_info'
        sname = base + chosen_stock + ext
        
        with open(sname, mode='a') as stock_file:
            stock_writer = csv.writer(stock_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
            stock_writer.writerow(['Date','Price','Pop'])
            stock_writer.writerows(zwr)
        
        hist_pop = []
        hist_price = []
        hist_date = []

    time.sleep(15)
```
